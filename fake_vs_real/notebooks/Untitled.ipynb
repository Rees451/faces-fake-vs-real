```python
    def _pair_generator(list_1, list_2):
        for fake_img, real_img in zip(list_1, list_2):
            fake_on = np.random.choice(['left', 'right'])

            if fake_on == 'left':
                yield fake_img, real_img, fake_on
            else:
                yield real_img, fake_img, fake_on
```

```python
class A():
    def __init__(self):
        print("Entering A")
        print("Leaving A")
        
    def method_A(self):
        print('method in a')

class B():
    def __init__(self):
        print("Entering B")
        print("Leaving B")

class C(A, B):
    def __init__(self):
        print("Entering C")
        A.__init__(self)
        B.__init__(self)
        print("Leaving C")
    def method_C(self):
        print('method in c')
```

```python
c = C()
```

```python
c.method_A()
```
